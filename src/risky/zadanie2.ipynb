В пункте 2 найти x_1, x_2: σ = 0 и соотвествующее E.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def eq(E_1, E_2, sigma_1, sigma_2, rho, E, sigma):
    return -(E_1-E_2)**2*sigma**2 + ((sigma_1-sigma_2)**2+2*(1-rho)*sigma_1*sigma_2)*E**2 + 2*((E_1+E_2)*rho*sigma_1*sigma_2-E_2*sigma_1**2-E_1*sigma_2**2)*E+(E_2**2*sigma_1**2+E_1**2*sigma_2**2-2*E_1*E_2*rho*sigma_1*sigma_2)
```

```python
E_values = np.linspace(0, 10, 100)
sigma_values = np.linspace(0, 10, 100)
E_grid, sigma_grid = np.meshgrid(E_values, sigma_values)
```

```python
def get_result(E_1, E_2, sigma_1, sigma_2, rho, E_grid, sigma_grid):
    # Вычисляем значения уравнения на сетке
    Z = eq(E_1, E_2, sigma_1, sigma_2, rho, E_grid, sigma_grid)

    # Строим график
    plt.figure(figsize=(10, 6))
    contour = plt.contour(sigma_grid, E_grid, Z, levels=[0], colors='blue')
    plt.clabel(contour, inline=True, fontsize=10)

    plt.scatter(sigma_1, E_1, color='red', s=50, label=f'E₁={E_1}, σ₁={sigma_1}', zorder=5)
    plt.scatter( sigma_2, E_2, color='red', s=50, label=f'E₂={E_2}, σ₂={sigma_2}', zorder=5)

    plt.text(sigma_1 + 0.25, E_1 + 0.1, f'E_1', 
            fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

    plt.text(sigma_2 + 0.25, E_2 + 0.1, f'E_2', 
            fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

    plt.ylabel('E (ожидаемая доходность)')
    plt.xlabel('σ (риск)')
    plt.title(f'Кривая\nE₁={E_1}, E₂={E_2}, σ₁={sigma_1}, σ₂={sigma_2}, ρ={rho}')

    plt.xticks(np.arange(sigma_grid.min(), sigma_grid.max()+1, 1.0))
    plt.yticks(np.arange(E_grid.min(), E_grid.max()+1, 1.0))

    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
E_1 = 1
E_2 = 6
sigma_1 = 1
sigma_2 = 4
rho = 1
Z = get_result(E_1, E_2, sigma_1, sigma_2, rho, E_grid, sigma_grid)
```

```python
E_1 = 1
E_2 = 6
sigma_1 = 1
sigma_2 = 4
rho = -1
Z = get_result(E_1, E_2, sigma_1, sigma_2, rho, E_grid, sigma_grid)
```
